Notebook ini memandu menjalankan agen FAQ secara langsung tanpa Streamlit.

```python
# Set working directory ke root repo agar requirements.txt ditemukan
import os
from pathlib import Path

repo_root = Path.cwd()
req_path = repo_root / "requirements.txt"
if not req_path.exists():
    for parent in Path.cwd().parents:
        candidate = parent / "requirements.txt"
        if candidate.exists():
            repo_root = parent
            req_path = candidate
            os.chdir(repo_root)
            break

print("Repo root:", repo_root)
print("Requirements:", req_path)
```

```python
# (Opsional) Instal dependensi sekali saja
import subprocess, sys
subprocess.run([sys.executable, "-m", "pip", "install", "-r", str(req_path)], check=False)
```

```python
# Ajukan pertanyaan contoh ke agen FAQ
from langchain_core.messages import HumanMessage

question = "Bagaimana cara menghubungi customer service Dexa Medica?"
state = {
    "messages": [HumanMessage(content=question)],
    "user_question": question,
    "attempt": 0,
    "rewritten_question": "",
}
result = graph.invoke(state, config={"configurable": {"thread_id": "faq-notebook-demo"}})
print(result["messages"][-1].content)
```
